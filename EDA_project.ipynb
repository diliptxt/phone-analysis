```
import seaborn as sns

data['hour'] = pd.to_datetime(data['date']).dt.hour

activity_heatmap = data.pivot_table(index='hour', columns='network', values='duration', aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(activity_heatmap, cmap='crest', annot=True, fmt='.1f', linewidths=0.5, linecolor='gray', cbar_kws={'label': 'Total Call Duration'})
plt.title('Call Activity by Hour and Network', fontsize=16)
plt.xlabel('Network', fontsize=12)
plt.ylabel('Hour of the Day', fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```


**Why Did You Pick the Specific Chart (Heatmap)?**

I chose a heatmap because it visually shows the call activity across different networks and hours of the day. The color intensity helps you quickly see:

High activity (darker colors) and low activity (lighter colors) for each network at different times.
